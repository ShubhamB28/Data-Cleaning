```python
from colour import Color
```

```python
def check_color(color):
    try: 
        Color(color) 
        return True 
    except ValueError:
        return False
```

```python

for row in df.index:
    _color = [word for word in df.title_long[row].split(' ') if check_color(word)]
    if len(_color)==0:
        df.color[row] = None
    else:
        df.color[row]= ' - '.join(_color)


  
```

```python
ccolor=['Morpinch','Cream','Turmeric','Biege','Brinjal', 'Burgundy','Multy',
        'Golden','Chikoo','Coffee','Copper','Light pista','Mustard','Pista','Wine','Peach',
        'Cookie Dough','Off-White','Dark','Rosewood','Rose','Firozi','firozi','teal','Sky','sky','Off-white','off-white','parrot','Parrot','Marron','Pista','pista','Lovender','Cookie','Multicolor','multicolor','(Grey)','(Black)','Jasper','(Navy)','(Maroon)','(Beige)','(Denim)'
,'Black-Red','Black-Grey','Graphite','Cobalt','Ash','Grey-Black','All-Black',
'Treebark','Tofu','Slate','Pepper','Marine','multi','Multi','Sapphire','Creame','Rama','rama','cookie']
```

```python
index = list(df[df.color.isnull()].index) #picking the index values of null rows
```

```python

for row in index:
    colorit=[]
    for word in df.title_long[row].split():
        if word in ccolor:
            colorit.append(word)  
    if len(colorit)>0:
        df.color[row] = colorit
    else:
        df.color[row]= None
df.head()
```
